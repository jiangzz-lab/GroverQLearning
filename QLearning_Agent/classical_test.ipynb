```python
import gym
import numpy as np
import matplotlib.pyplot as plt
```

```python
from QQL_learner_trainer import ClassicalLearner
# test
# choose env
envtest = gym.make("FrozenLake-v1", is_slippery=False, render_mode="ansi")
# init learner
Elliot_c = ClassicalLearner(envtest)
# good hyperparms (hand-tuned)
hyperp = {'alpha': 0.1,
          'gamma': 0.99,
          'eps': 0.0,
          'max_epochs': 3000,
          'max_steps': 15}
# set hyperparms
Elliot_c.set_hyperparams(hyperp)

# TRAIN
steps_in_all_epochs,target_reached_in_all_epochs,_ = Elliot_c.train()

plt.plot(steps_in_all_epochs)
```

```python
print(steps_in_all_epochs[-1])
plt.scatter(range(len(target_reached_in_all_epochs)),target_reached_in_all_epochs)
```
