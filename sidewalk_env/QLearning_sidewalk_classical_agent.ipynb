```python
import matplotlib.pyplot as plt
import numpy as np
from SideWalkEnv import *
import sys
sys.path.append('..\QLearning_Agent')
```

```python
env_with_obstacle_train = side_walk_env_with_obstacle(50,15,12,2,0.2)
env_with_obstacle_train.plot_roadmap()
```

```python
from QQL_learner_trainer import ClassicalLearner

# env = side_walk_env_with_obstacle(50,15,12,2,0.2)
# env.plot_roadmap()
Elliot_obstacle_cl = ClassicalLearner(env_with_obstacle_train,env_type='local')
hyperp = {'alpha': 0.1,
          'gamma': 0.8,
          'eps': 0.2,
          'max_epochs': 5000,
          'max_steps': 1000}

# set hyperparms
Elliot_obstacle_cl.set_hyperparams(hyperp)
# TRAIN
steps_in_all_epochs,target_reached_in_all_epochs,_ = Elliot_obstacle_cl.train()

# plt.plot(steps_in_all_epochs)
print(Elliot_obstacle_cl.Q_values)
```

```python
env_with_obstacle_test = side_walk_env_with_obstacle(50,15,12,2,0.15)
trajectory_obstacle = env_with_obstacle_test.trajectory(Elliot_obstacle_cl.Q_values)
n_try = 1
while trajectory_obstacle[0][-1] != env_with_obstacle_test.nx-1 and n_try < 100:
    trajectory_obstacle = env_with_obstacle_test.trajectory(Elliot_obstacle_cl.Q_values)
env_with_obstacle_test.plot_roadmap_with_trajectory('avoiding obstacles_Classical agent',trajectory_obstacle)
# print(trajectory[0])
```
