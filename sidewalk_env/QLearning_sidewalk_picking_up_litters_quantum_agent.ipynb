```python
import matplotlib.pyplot as plt
import numpy as np
from SideWalkEnv import *
import sys
sys.path.append('..\QLearning_Agent')
```

```python
env_with_litter_train = side_walk_env_with_litter(50,15,12,2,0.25)
env_with_litter_train.plot_roadmap()
```

```python
from QQL_learner_trainer import GroverQlearner
Elliot_litter = GroverQlearner(env_with_litter_train,env_type='local')
hyperp = {'k': 0.1,
          'alpha': 0.1,
          'gamma': 0.8,
          'eps': 0.01,
          'max_epochs': 800,
          'max_steps': 300}

# set hyperparms
Elliot_litter.set_hyperparams(hyperp)
# TRAIN
steps_in_all_epochs,target_reached_in_all_epochs,_ = Elliot_litter.train()
```

```python
env_with_litter_test = side_walk_env_with_litter(50,15,12,2,0.15)
trajectory_litter = env_with_litter_test.trajectory(Elliot_litter.Q_values)
env_with_litter_test.plot_roadmap_with_trajectory('picking up litters_Quantum agent',trajectory_litter)
```

```python
print(Elliot_litter.Q_values)
```
